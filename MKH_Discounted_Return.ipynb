```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("dark")
```

```python
def simulate(num_steps):
    gammas = np.linspace(0, 1, num=101)

    def policy_a(gamma):
        G_0 = 0
        for t in range(1, num_steps, 2):
            G_0 += np.power(gamma, t-1)

        return G_0

    def policy_b(gamma):
        G_0 = 0
        for t in range(2, num_steps, 2):
            G_0 += 2*np.power(gamma, t-1)

        return G_0

    policy_a = np.vectorize(policy_a)
    rewards_a = policy_a(gammas)

    policy_b = np.vectorize(policy_b)
    rewards_b = policy_b(gammas)

    fig, ax = plt.subplots()
    sns.lineplot(x=gammas, y=rewards_a, label="Policy A")
    sns.lineplot(x=gammas, y=rewards_b, label="Policy B")
    plt.xlabel("Discount Factor")
    plt.ylabel("Total Reward")
    plt.title(f"Policy Simulations over {num_steps} steps")
    plt.grid()
    sns.despine()
    plt.show()
```

```python
simulate(20)
```
